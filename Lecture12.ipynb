```python
import qiskit
from qiskit import QuantumCircuit
q=QuantumCircuit(10)
q.x(0)
q.x(1)
q.x(2)
q.x(3)
q.x(4)
q.x(5)
q.x(6)
q.x(7)
q.x(8)
q.x(9)
for i in range(10):
    q.y(i)
for i in range(10):
    q.z(i)
for i in range(9):
    q.cx(i,i+1)
q.draw()
```

```python
from qiskit import QuantumCircuit
from qiskit.circuit.library import HGate, CXGate

qc = QuantumCircuit(2)

my_h = HGate()
my_cx = CXGate()

qc.append(my_h, [0])
qc.append(my_cx, [0, 1])

qc.draw("text")
```
